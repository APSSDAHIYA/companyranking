```python
#reading the file
import pandas as pd
import numpy as np

rawfile = open("fortune1000.txt","r")

print(rawfile)

alldata = rawfile.readlines()
#print(alldata)

rawdatalist =[line.strip() for line in alldata]

####
rawarray = np.array(rawdatalist)
print(rawarray)

#Rank, Name, Revenue($m), % Change in Revenue, Profits($m), % Change in Profits, Assets($m),
#Market Value($m), Change in Rank (1000), Employees, and Change in Rank (500 Only)
df2 =pd.DataFrame(rawarray.reshape(1001,11), columns=["Rank","Name", "Revanue","% Change in Revenue",
                                                      "Profits","% Change in Profits", "Assets",
                                                      "Market_Value","Change in Rank(1000)",
                                                      "Employees","Change in Rank (500 Only)"])

#print(df2)

df3 =df2.drop(1000)

#print(df3)
#print(rawdatalist)--------

#Raw_data_frame = pd.DataFrame(rawdatalist)

#print(Raw_data_frame)

###
#df1 =np.reshape(Raw_data_frame,(1001,11))
#df2 =pd.DataFrame(Raw_data_frame.reshape(1001,11))
#df1 = Raw_data_frame.reshape(1001,11)

#print(df2)

final_data_frame = df3.drop(['% Change in Revenue','% Change in Profits',
                             'Change in Rank(1000)','Change in Rank (500 Only)'],axis=1)
#print(final_data_frame)

final_data_frame['Revenue'] = final_data_frame.Revanue.str.replace('$','')
final_data_frame['Revanue'] = final_data_frame.Revanue.str.replace(',','')

final_data_frame['Profits'] = final_data_frame.Profits.str.replace('$','')
final_data_frame['Profits'] = final_data_frame.Profits.str.replace(',','')

final_data_frame['Assets'] = final_data_frame.Assets.str.replace('$','')
final_data_frame['Assets'] = final_data_frame.Assets.str.replace(',','')


final_data_frame['Market_Value'] = final_data_frame.Market_Value.str.replace('$','')

final_data_frame['Market_Value'] = final_data_frame.Market_Value.str.replace(',','')

final_data_frame['Employees'] = final_data_frame.Employees.str.replace(',','')

#final_data_frame[]


print(final_data_frame.tail())

final_data_frame.to_csv('final_data_frame3.csv')
```
